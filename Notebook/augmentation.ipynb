```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("book.jpeg")
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

# Horizontal Flip

```python
from torchvision import transforms

transform = transforms.RandomHorizontalFlip()

new_img = transform(img)
plt.imshow(new_img)
plt.title("Flipped Image")
plt.axis("off")
plt.show()
```

# Vertical Flip

```python
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms

img = Image.open("book.jpeg")

transform = transforms.RandomVerticalFlip(p=1)
new_img = transform(img)

output_dir = Path("augmented_images")
output_dir.mkdir(parents=True, exist_ok=True)

output_path = output_dir / "book_vertical_flip.jpeg"
new_img.save(output_path)
print(f"Saved flipped image as {output_path}")

plt.imshow(new_img)
plt.title("Vertical Flip")
plt.axis("off")
plt.show()
```

# More detail

```python
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("book.jpeg")

transform = transforms.RandomHorizontalFlip(p=0.1)

for i in range(5):
    new_img = transform(img)
    plt.figure()
    plt.imshow(new_img)
    plt.title(f"Trial {i+1}")
    plt.axis("off")
    plt.show()
```

# 4. Color Jitter( Changes brightness, contrast, saturation, and hue.)


```python
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms

img = Image.open("book.jpeg")

transform = transforms.ColorJitter(
    brightness=0.9, #kattasi xiraroq
    contrast=0.8,   # tiniqni tiniq, xirani xira
    saturation=0.6,  # kuchli rangni yanada kuchli qiladi
    hue=0.3     # Yaqin ragnga o'zgartiradi
)
new_img = transform(img)

plt.imshow(new_img)
plt.title("Color Jitter")
plt.axis("off")
plt.show()
```

# Random Affine

```python
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms

img = Image.open("book.jpeg")

transform = transforms.RandomAffine(
    degrees=30, # rotation (-30,30)
    translate=(0.1, 0.1),   # horizontally and verticaly
    scale=(0.8, 1.9),  # zoom
    shear=10 #(-10,10)  # qiyalashtirish(yotkizish)
)
new_img = transform(img)

plt.imshow(new_img)
plt.title("Affine Transform")
plt.axis("off")
plt.show()
```

# Random Grayscale

```python
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms

img = Image.open("book.jpeg")

transform = transforms.RandomGrayscale(p=0.1)
new_img = transform(img)

plt.imshow(new_img, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

# Combined Augmentations

```python
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms

img = Image.open("book.jpeg")

transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(30),
    transforms.ColorJitter(brightness=0.3, contrast=0.3),
    transforms.RandomCrop((150 , 150))
])

new_img = transform(img)

plt.imshow(new_img)
plt.title("Combined Augmentations")
plt.axis("off")
plt.show()
```
