```python
import sys
sys.path.append(r"C:\Users\Rasulbekk\Desktop\Practise\Source")  

from auto_pipeline import Auto_Pipeline
import pandas as pd
from sklearn.ensemble import RandomForestClassifier


file_path = r"C:\Users\Rasulbekk\Desktop\Practise\Data\train.csv"
df = pd.read_csv(file_path)


df = df[df['Survived'].notna()]


titanic_pipeline = Auto_Pipeline(
    df=df,
    target='Survived',
    model=RandomForestClassifier(n_estimators=100, random_state=42)
)


titanic_pipeline.fit()


preds = titanic_pipeline.predict()
print("Predictions:", preds[:50])  


score = titanic_pipeline.score()
print("Accuracy:", score)
```
