# Preprocessing= Transforms

```python
from PIL import Image
import matplotlib.pyplot as plt

samples = [
    ("anor.jpg", 0),
    ("uzum.jpg", 1),
    ("banan.jpg", 2),
    ("olma.jpg", 3)
]

image_file, label = samples[0]

img = Image.open(image_file)
resized_img = img.resize((128, 128))

print("Image file:", image_file)
print("Label:", label)
print("Original size:", img.size)
print("Resized size:", resized_img.size)

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title(f"Original\n{img.size}")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(resized_img)
plt.title(f"Resized\n{resized_img.size}")
plt.axis("off")

plt.show()
```

# Brightness (yorqinlikni o‘zgartirish)

```python
from PIL import ImageEnhance

enhancer = ImageEnhance.Brightness(img)
bright_img = enhancer.enhance(1.5)  # >1 bright, <1 dark

plt.imshow(bright_img)
plt.title("Brightness Increased")
plt.axis("off")
plt.show()
```

# Contrast

```python
enhancer = ImageEnhance.Contrast(img)
contrast_img = enhancer.enhance(2.0)

plt.imshow(contrast_img)
plt.title("High Contrast")
plt.axis("off")
plt.show()
```

# Color Saturation

```python
enhancer = ImageEnhance.Color(img)
color_img = enhancer.enhance(2.0)

plt.imshow(color_img)
plt.title("High Saturation")
plt.axis("off")
plt.show()
```

# Blur (xiralashtirish)

```python
from PIL import ImageFilter

blur_img = img.filter(ImageFilter.BLUR)

plt.imshow(blur_img)
plt.title("Blur")
plt.axis("off")
plt.show()
```

# Sharpen (aniqlashtirish)

```python
sharp_img = img.filter(ImageFilter.SHARPEN)

plt.imshow(sharp_img)
plt.title("Sharpen")
plt.axis("off")
plt.show()
```

# Edge Detection (chegaralarni topish)

```python
edge_img = img.filter(ImageFilter.FIND_EDGES)

plt.imshow(edge_img)
plt.title("Edges")
plt.axis("off")
plt.show()
```

# Random Crop (tasodifiy kesish)

```python
import random

w, h = img.size
left = random.randint(0, w//4)
top = random.randint(0, h//4)
right = left + 100
bottom = top + 100

random_crop = img.crop((left, top, right, bottom))

plt.imshow(random_crop)
plt.title("Random Crop")
plt.axis("off")
plt.show()
```

# Padding (atrofini kengaytirish)

```python
import numpy as np

img_array = np.array(img)
padded = np.pad(img_array, ((20,20),(20,20),(0,0)), mode='constant')

plt.imshow(padded)
plt.title("Padding")
plt.axis("off")
plt.show()
```

# Affine Transform (siljitish)

```python
affine_img = img.transform(
    img.size,
    Image.AFFINE,
    (1, 0, 50, 0, 1, 30)  # shift x=50, y=30
)

plt.imshow(affine_img)
plt.title("Affine Transform (Shift)")
plt.axis("off")
plt.show()
```

# Gaussian Noise qo‘shish

```python
import numpy as np

img_array = np.array(img).astype(np.float32)

noise = np.random.normal(0, 25, img_array.shape)
noisy_img = img_array + noise

noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)

plt.imshow(noisy_img)
plt.title("Gaussian Noise")
plt.axis("off")
plt.show()
```
